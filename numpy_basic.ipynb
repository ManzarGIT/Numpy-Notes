# Numpy Basic

```python
import numpy as np
a=np.array([1,2,3,4])
print(a)
```

```python
#2d array
b=np.array([[1,2,3,4],[5,6,7,8]])
print(b)
```

```python
#3d array
c=np.array([[[1,2,3,4],[5,6,7,8]]])
print(c)
```

```python
#dtype array
d=np.array([1,2,3,4],dtype=complex)
print(d)
d1=np.array([1,2,3,4],dtype=float)
print(d1)
d2=np.array([1,2,3,4],dtype=bool)
print(d2)
```

```python
#arange
e=np.arange(1,5)
print(e)
e1=np.arange(1,5,2)
print(e1)
```

```python
#reshape
f=np.array([[1,2,3,4],[5,6,7,8]])
print(f.reshape(4,2))


print(f.reshape(2,4))
```

```python
#zeros and ones
g=np.zeros((3,4))
print(g)
h=np.ones((3,4))
print(h)
```

```python
#random
# it will give you random number betwwen 0 to 1
i=np.random.random((3,4))
print(i)
```

```python
#linspace
j=np.linspace(1,5,10)
print(j)
```

```python
#identity
k=np.eye(3)
print(k)

k1=np.identity(5)
print(k1)
```

# Array Attributes

```python
l1=np.arange(10)
l2=np.arange(12,dtype=float).reshape(3,4)
l3=np.arange(8).reshape(2,2,2)
print(l1)
print(l2)
print(l3)
```

```python
#ndim
print(l1.ndim)
print(l2.ndim)
print(l3.ndim)
```

```python
#shape
print(l1.shape)
print(l2.shape)
print(l3.shape)
```

```python
#size
# The size attribute of a NumPy array returns the total number of elements in the array.
print("Size of l1:", l1.size)
print("Size of l2:", l2.size)
print("Size of l3:", l3.size)
```

```python
#itemsize
# The itemsize attribute of a NumPy array returns the size of each element in the array in bytes.
print("Itemsize of l1:", l1.itemsize)
print("Itemsize of l2:", l2.itemsize)
```

```python
#dtype
print(l1.dtype)
print(l2.dtype)
print(l3.dtype)
```

# Stacking

```python
#horizontal stacking
a1=np.arange(12).reshape(3,4)
a2=np.arange(12,24).reshape(3,4)
a3=np.arange(24,36).reshape(3,4)

print(a1)
print(a2)
np.hstack((a1,a2))
np.hstack((a1,a2,a3))
```

```python
#vertical Stacking

np.vstack((a1,a2))
```
